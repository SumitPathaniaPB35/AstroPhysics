```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file='https://drive.google.com/uc?id=1BQVc6MHjQFtDC9iP1isT_K4ojVe_Oil-'
star_data=pd.read_csv(file)
star_data.info()
```

Observations
1) Dataset consists of 240 rows, 6 feature and 1 target columns,

  Absolute Temperature (in K)
  Relative Luminosity (L/Lo)
  Relative Radius (R/Ro)
  Absolute Magnitude (Mv)
  Star Color
  Spectral Class
  Star Type (Target classes)
  Here,

  Lo = 3.828 x 10^26 Watts (Avg Luminosity of Sun)
  Ro = 6.9551 x 10^8 m (Avg Radius of Sun)


2) Two categorical features (object type) which will need some kind of encoding,

    Star Color
    Spectral Class


3) It consists of some different features of stars. Information on the star type is given below,

   0 → Brown Dwarf
   1 → Red Dwarf
   2 → White Dwarf
   3 → Main Sequence
   4 → Supergiants
   5 → Hypergiants

```python
#Creating Subplots of boxplot 
plt.figure(figsize=(20,8))
plt.style.use('default')
plt.suptitle('Visualization the outliers in numeric features of Star Type',weight='bold',fontsize=15)
for i in range(4):
    plt.subplot(1,4,i+1)
    sns.boxplot(x=star_data['Star type'],y=star_data.iloc[:,i],palette='terrain_r')
    plt.title(star_data.columns[i],color='red')
    plt.savefig('boxplot_star_type.png')
    plt.ylabel('')

```

```python
def line_subplot(star_data,colors,i):
    plt.subplot(4,1,i+1)
    plt.plot(star_data.iloc[:,i],color=colors[i])
    plt.title(star_data.columns[i],color='red') 
```

```python
plt.figure(figsize=(12,8))
plt.suptitle('Visualization the distribution of Numeric Features\n',weight='bold',fontsize=17)
colors=['violet','cyan','lime','magenta']
for i in range(4):
    line_subplot(star_data,colors,i)

plt.savefig('line_subplot.png')
#automatically adjusts the white spaces in the figure
plt.tight_layout()
plt.show()
```

HR Diagram

```python
star_data.head()
```

```python
# Python dictionaries -->To be used to get properties of star_type

star_type=star_data['Star type'].values
star_types = {
    0: {'label': 'Brown Dwarf', 'color': 'brown', 'size': 30, 'marker': '.'},
    1: {'label': 'Red Dwarf', 'color': 'red', 'size': 35, 'marker': '.'},
    2: {'label': 'White Dwarf', 'color': 'white', 'size': 40, 'marker': '.'},
    3: {'label': 'Main Sequence', 'color': 'cyan', 'size': 30, 'marker': 'o'},
    4: {'label': 'Supergiants', 'color': 'orange', 'size': 100, 'marker': 'o'},
    5: {'label': 'Hypergiants', 'color': 'maroon', 'size': 150, 'marker': 'o'}
}
```

```python
temp=star_data['Temperature (K)'].values
abs_mag=star_data['Absolute magnitude(Mv)'].values

plt.scatter(temp,abs_mag)


#  get current axis

plt.gca().invert_xaxis() #Desc order axis
plt.gca().invert_yaxis()

plt.show()
```

```python
len(star_type)
```

```python
axes=[]
labels=set()
plt.figure(figsize=(10,6))
plt.title(f'Hertzsprung - Russell Diagram for {len(star_type)} Stars',color='violet',fontsize=16,weight='bold')
for i in range(len(star_type)):
    properties=star_types[star_type[i]]
    if properties['label'] not in labels:
        ax=plt.scatter(temp[i],abs_mag[i],s=properties['size'],c=properties['color'],marker=properties['marker'],label=properties['label'])
        axes.append(ax)
        labels.add(properties['label'])
    else:
        plt.scatter(temp[i],abs_mag[i],s=properties['size'],c=properties['color'],marker=properties['marker'],label=properties['label'])
#Adding data for sun
ax_sun=plt.scatter(5778,4.83,s=75,c='yellow',marker='o',label='Sun')
axes.append(ax_sun)
labels.add('Sun')
plt.legend(handles=axes)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.savefig('hr_diagram.png')
plt.show()
```
